        <p>Boats are smaller vessels that serve a wide range of purposes, from recreational activities to specialized professional tasks. They include:</p>
        <ul>
            <li><strong>Fishing boats:</strong> Essential for the fishing industry, these boats vary in size and design, from small trawlers to large factory ships.</li>

            <li><strong>Speedboats:</strong> Designed for high speed, they are used in racing, rescue missions, and recreational activities.</li>
            <li><strong>Ferries:</strong> Transport people and vehicles across short distances, playing a vital role in local transportation networks.</li>

        <p>Offshore structures, including oil and gas platforms, wind farms, and research stations, allow us to harness the vast resources of the ocean. These structures are engineering feats, designed to withstand harsh marine environments while providing a stable base for various operations.</p>
        <h3>Types of Offshore Structures</h3>
        <ul>
            <li><strong>Fixed Platforms:</strong> These structures are anchored directly to the seabed and are typically used in shallow waters. They include jacket platforms and gravity-based structures.</li>
            <li><strong>Floating Platforms:</strong> Used in deeper waters, these platforms are not fixed to the seabed. Types include semi-submersibles, FPSOs (Floating Production, Storage, and Offloading units), and TLPs (Tension Leg Platforms).</li>
            <li><strong>Subsea Systems:</strong> These are installed on the ocean floor and are used for tasks such as oil and gas extraction, where the surface infrastructure is not viable.</li>
        </ul>

            <li><strong>Safety:</strong> Implementing safety features and protocols to protect the crew, cargo, and environment.</li>
